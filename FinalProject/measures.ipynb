```python
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
cpu1 = [i.strip().split() for i in open("source time.dat").readlines()]
times_default = np.array(cpu1)
times_default = times_default.reshape(times_default.shape[0])
times = np.zeros_like(times_default, dtype=float)
for i in range(len(times_default)):
    times[i] = float(times_default[i])

cpu2 = [i.strip().split() for i in open("parallel time.dat").readlines()]
times_default = np.array(cpu2)
times_default = times_default.reshape(times_default.shape[0])
times_parallel = np.zeros_like(times_default, dtype=float)
for i in range(len(times_default)):
    times_parallel[i] = float(times_default[i])
```

```python
times = times[:len(times_parallel)]

plt.figure(figsize=(15,15))
plt.title("Default RRT")
plt.plot(times)
plt.plot(times_parallel)
plt.legend(["default", "non-blocking"])
plt.ylabel("Time")
plt.xlabel("Runs")
plt.grid()
plt.show()

print("Mean time default RRT:", round(np.mean(times), 5))
```
